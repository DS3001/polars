# Visualization

```python
import numpy as np

tdf = df.with_columns(
    pl.col("BondAmountAtInitialContact").cast(pl.Float32,strict=False).alias("bond")
)
tdf = tdf.with_columns(
    pl.col("bond").log().alias('log_bond')
)
#tdf['log_bond'] = np.log( tdf['bond'] )

tdf['bond'].describe()
```

```python
tdf['Defendant_AgeGroup']
```

```python
import plotly.express as px

#px.bar( df = tdf, x = 'Defendant_AgeGroup', )

fig = px.histogram(tdf,x='log_bond', nbins = 50, width = 500)
fig.show()

fig = px.box(tdf, y = 'log_bond', width = 500)
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(tdf['log_bond'],bins=30)
plt.show()

sns.kdeplot(tdf['log_bond'])
plt.show()

sns.boxplot(tdf['log_bond'])
plt.show()
```
